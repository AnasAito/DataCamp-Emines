```python
s = "HelLo WorLd!!"

# normalize characters
s = s.lower()

# where to store count chars
dict_characters = {}

for current_char in s:
    # check if already there
    if current_char in dict_characters.keys():
        continue
    
    # count occurences
    dict_characters[current_char] = 0
    for c in s:
        # increment count
        if c == current_char:
            dict_characters[current_char] += 1


# result
print(dict_characters)
```

```python
import numpy as np

# create random matrix
M = np.random.uniform(low=-1, high=1, size=(5, 6))

print("initial matrix \n {}".format(M))

# operation
for i in range(0, 5, 2):
    M[:, i] -= 2 * M[:, i+1]

# print result
print("final matrix \n {}".format(M))
```

```python
# neg value to zero
M = M * (M >= 0)

print("non negative \n {}".format(M))

# mean rows
mean = np.mean(M, axis=1)

M -= mean[:, None]

print("M minus mean \n {}".format(M))
```

```python
import numpy as np

M = np.random.uniform(low=-1, high=1, size=(5, 10))

G = M.T @ M
eigen_val_G, _ = np.linalg.eig(G)

# semi definte positive
print("G is semi-definte positive: {0}".format(all(eigen_val_G > 0)))

# rank
print("rank of G: {}".format(np.linalg.matrix_rank(G)))

# euclidean norm
print("euclidean norm of G: {}".format(np.linalg.norm(G)))

# Schatten 2-norm
print("Schatten 2-norm of G: {}".format(np.linalg.norm(G, ord="nuc")))

# std of cols
std_G = np.std(G, axis=0)
print("std of cols: \n {}".format(std_G))
```

```python
import plotly.graph_objects as go

# params
interval = [-1, 2]
nb_points = 100
x = np.linspace(*interval, nb_points)

# plots
fig = go.Figure(data=[
    go.Scatter(
        x=x,
        y=x ** d,
        mode="lines",
        name=f"x^{d}"
    )
for d in [2, 3, 4]])


fig.update_layout(
    title="Plots x^d for d in {2, 3, 4}",
    xaxis_title="x",
    yaxis_title="y",
)

fig.show()
```
