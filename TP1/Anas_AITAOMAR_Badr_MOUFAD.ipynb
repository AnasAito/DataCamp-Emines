# question 4

```python
s = "HelLo WorLd!!"

# normalize characters
s = s.lower()

# where to store count chars
dict_characters = {}

for current_char in s:
    # check if already there
    if current_char in dict_characters.keys():
        continue
    
    # count occurences
    dict_characters[current_char] = 0
    for c in s:
        # increment count
        if c == current_char:
            dict_characters[current_char] += 1


# result
print(dict_characters)
```

```python

def serie_term(k):
    sqr_k = k**2
    return 4 * sqr_k /( 4 * sqr_k -1)
```

```python
%%timeit
# using loop 

prod_loop = 2 
for k in range(1,1000000):
    prod_loop = prod_loop * serie_term(k)
```

```python
%%timeit
# compute a product serie using numpy 

x = np.arange(1,1000000)
prod_numpy = 2*np.prod(serie_term(x))

```

# question 13

```python
M = np.random.uniform(low=-1, high=1, size=(5, 10))

G = M.T @ M
eigen_val_G, _ = np.linalg.eig(G)

# semi definte positive
print("G is semi-definte positive: {0}".format(all(eigen_val_G > 0)))

# rank
print("rank of G: {}".format(np.linalg.matrix_rank(G)))

# euclidean norm
print("euclidean norm of G: {}".format(np.linalg.norm(G)))

# Schatten 2-norm
print("Schatten 2-norm of G: {}".format(np.linalg.norm(G, ord="nuc")))

# std of cols
std_G = np.std(G, axis=0)
print("std of cols: \n {}".format(std_G))
```
